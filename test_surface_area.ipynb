```python
from pdb_parser import parse
from shrake_rupley import calculate_surface_area, parse_results_by_residue
```

```python
atoms = parse("1ubq")
print("Number of atoms in PDB file: {:,}".format(len(atoms)))
```

```python
results = calculate_surface_area(atoms)
parsed_results = parse_results_by_residue(results)
print("Total surface area: {:.2f} A^2".format(sum(parsed_results.area)))
```
